```python
# Libraries
import pandas as pd
import matplotlib as mp
```

```python
dirData = 'data/'
leapfrogDirty = pd.read_csv(dirData + 'external/majorcourses.txt', sep='|', 
                            header=None, names=['class', 'degree', 'degreeDesc'])

leapfrogDirty.head()
```

```python
# add more broad roll up categories

# break class apart
leapfrogDirty['classMajor'] = leapfrogDirty['class'].apply(lambda x: x.split(' ')[0])
leapfrogDirty['classNum'] = leapfrogDirty['class'].apply(lambda x: x.split(' ')[1])

# Break degree apart
leapfrogDirty['degreeMajor'] = leapfrogDirty['degree'].apply(lambda x: x.split('-')[0])
# reverse the string
leapfrogDirty['degreeType'] = leapfrogDirty['degree'].apply(lambda x: x.split('-')[1][::-1])

# actually save that too
#leapfrogDirty['degreeDesc'] = leapfrogDirty['degreeDesc'].apply(lambda x: x.split(' ')[0] + ' ' + x.split(' ')[2][::-1])


#Human readable degree
leapfrogDirty['degreeMajorRead'] = leapfrogDirty['degreeDesc'].apply(lambda x: x.split(' -')[0])

# conditionally catch the text between parantheses
leapfrogDirty['degreeFocus'] = leapfrogDirty['degreeDesc'].apply(lambda x: x[x.find("(")+1:x.find(")")] if '(' in x else '')

leapfrogDirty.iloc[150:175]
```

```python
# get list of all majors
major = leapfrogDirty[['degreeMajor', 'degreeMajorRead']]

for major in major['degreeMajorRead'].unique():
    print (major)
```

```python
# write it out to csv
leapfrogDirty.to_csv((dirData + 'processed/leapfrog-clean.csv'), index=False)
```
